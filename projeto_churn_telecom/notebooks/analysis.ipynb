# 📊 Análise de Churn - Telecom Company

Este notebook contém a análise completa de churn para uma empresa de telecomunicações.

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
```

## 1. Carregar Dados

```python
# Carregar dados
df = pd.read_csv('../data/raw/telco_churn.csv')
print(f"Dataset carregado: {df.shape[0]} linhas, {df.shape[1]} colunas")
```

## 2. Limpeza de Dados

```python
# Converter TotalCharges para numérico
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(0, inplace=True)

# Criar features
df['TenureGroup'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 72], labels=['0-12', '12-24', '24-48', '48-72'])
print("Dados limpos com sucesso!")
```

## 3. Análise de Churn

```python
# Taxa de churn por contrato
churn_contract = df.groupby('Contract')['Churn'].mean() * 100
print("Taxa de Churn por Contrato:")
print(churn_contract.sort_values(ascending=False))
```

```python
# Plotar gráfico
plt.figure(figsize=(10, 6))
churn_contract.plot(kind='bar', color=['#e74c3c', '#f39c12', '#2ecc71'])
plt.title('Taxa de Churn por Tipo de Contrato')
plt.ylabel('Taxa de Churn (%)')
plt.xticks(rotation=0)
plt.show()
```

## 4. Conclusões

- Contratos month-to-month têm maior taxa de churn
- Clientes novos (<12 meses) são mais propensos a cancelar
- Fiber optic apresenta desafios de qualidade
